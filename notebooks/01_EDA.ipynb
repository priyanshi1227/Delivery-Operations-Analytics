```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)

# Load Dataset

df = pd.read_csv("../data/raw/train.csv")

print("=" * 70)
print("Dataset Shape")
print(df.shape)

print("=" * 70)
print("Dataset Information")
print(df.info())

print("=" * 70)
print("First Five Records")
print(df.head())

print("=" * 70)
print("Summary Statistics")
print(df.describe(include="all"))

print("=" * 70)
print("Missing Values")
print(df.isnull().sum())

print("=" * 70)
print("Duplicate Rows")
print(df.duplicated().sum())

print("=" * 70)
print("Unique Values")
print(df.nunique())

# Clean delivery time
df["Time_taken(min)"] = (
    df["Time_taken(min)"]
      .astype(str)
      .str.extract("(\d+)")
      .astype(float)
)

# Numeric conversion
df["Delivery_person_Ratings"] = pd.to_numeric(
    df["Delivery_person_Ratings"],
    errors="coerce"
)

df["Delivery_person_Age"] = pd.to_numeric(
    df["Delivery_person_Age"],
    errors="coerce"
)

# Visualizations

plt.figure(figsize=(8,5))
plt.hist(df["Time_taken(min)"], bins=25)
plt.title("Delivery Time Distribution")
plt.xlabel("Minutes")
plt.ylabel("Orders")
plt.show()

plt.figure(figsize=(8,5))
df["Road_traffic_density"].value_counts().plot(kind="bar")
plt.title("Traffic Density")
plt.show()

plt.figure(figsize=(8,5))
df["Weatherconditions"].value_counts().plot(kind="bar")
plt.title("Weather Conditions")
plt.show()

plt.figure(figsize=(8,5))
df["Type_of_vehicle"].value_counts().plot(kind="bar")
plt.title("Vehicle Types")
plt.show()

plt.figure(figsize=(8,5))
df["City"].value_counts().plot(kind="bar")
plt.title("Orders by City")
plt.show()

plt.figure(figsize=(8,5))
plt.hist(df["Delivery_person_Ratings"], bins=20)
plt.title("Driver Ratings")
plt.show()

plt.figure(figsize=(8,5))
plt.scatter(
    df["Delivery_person_Ratings"],
    df["Time_taken(min)"]
)
plt.xlabel("Ratings")
plt.ylabel("Delivery Time")
plt.title("Ratings vs Delivery Time")
plt.show()

print("=" * 70)
print("EDA Completed Successfully")
```
