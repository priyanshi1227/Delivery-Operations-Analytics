```
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)

# Load Dataset

df = pd.read_csv("../data/processed/final_delivery_dataset.csv")

print("="*70)
print("DELIVERY OPERATIONS KPI DASHBOARD")
print("="*70)

# Overall KPIs

total_orders = len(df)

delivered_orders = len(
    df[df["Order_Status"]=="Delivered"]
)

cancelled_orders = len(
    df[df["Order_Status"]=="Cancelled"]
)

total_revenue = df["Revenue"].sum()

average_order_value = df["Order_Value"].mean()

average_delivery_time = df["Time_taken(min)"].mean()

average_rating = df["Delivery_person_Ratings"].mean()

average_tip = df["Tip"].mean()

delivery_fee = df["Delivery_Fee"].sum()

on_time = (
    len(df[df["On_Time"]=="Yes"])/len(df)
)*100

print(f"Total Orders : {total_orders:,}")
print(f"Delivered Orders : {delivered_orders:,}")
print(f"Cancelled Orders : {cancelled_orders:,}")
print(f"Revenue : ₹{total_revenue:,.2f}")
print(f"Average Order Value : ₹{average_order_value:.2f}")
print(f"Average Delivery Time : {average_delivery_time:.2f} mins")
print(f"Average Driver Rating : {average_rating:.2f}")
print(f"Average Tip : ₹{average_tip:.2f}")
print(f"Delivery Fee Revenue : ₹{delivery_fee:,.2f}")
print(f"On-Time Delivery : {on_time:.2f}%")

# Revenue by City

city_revenue = (
    df.groupby("City")
      .agg(
          Revenue=("Revenue","sum"),
          Orders=("ID","count")
      )
      .sort_values("Revenue",ascending=False)
)

print(city_revenue)

plt.figure(figsize=(10,5))
city_revenue["Revenue"].plot(kind="bar")
plt.title("Revenue by City")
plt.ylabel("Revenue")
plt.show()

# Revenue by Payment Method

payment = (
    df.groupby("Payment_Method")["Revenue"]
      .sum()
      .sort_values(ascending=False)
)

print(payment)

plt.figure(figsize=(8,5))
payment.plot(kind="bar")
plt.title("Revenue by Payment Method")
plt.show()

# Orders by Vehicle

vehicle = (
    df.groupby("Type_of_vehicle")
      .size()
      .sort_values(ascending=False)
)

print(vehicle)

plt.figure(figsize=(8,5))
vehicle.plot(kind="bar")
plt.title("Orders by Vehicle")
plt.show()

# Traffic Impact

traffic = (
    df.groupby("Road_traffic_density")
      .agg(
          Avg_Time=("Time_taken(min)","mean"),
          Orders=("ID","count")
      )
)

print(traffic)

# Weather Impact

weather = (
    df.groupby("Weatherconditions")
      .agg(
          Avg_Time=("Time_taken(min)","mean"),
          Revenue=("Revenue","sum")
      )
)

print(weather)

# Festival Analysis

festival = (
    df.groupby("Festival")
      .agg(
          Orders=("ID","count"),
          Revenue=("Revenue","sum"),
          Avg_Time=("Time_taken(min)","mean")
      )
)

print(festival)

# Peak Hour Analysis

peak = (
    df.groupby("Peak_Hour")
      .agg(
          Orders=("ID","count"),
          Revenue=("Revenue","sum")
      )
)

print(peak)

# Monthly Revenue

monthly = (
    df.groupby("Month_Name")
      .agg(
          Revenue=("Revenue","sum")
      )
)

print(monthly)

plt.figure(figsize=(10,5))
monthly["Revenue"].plot(kind="bar")
plt.title("Monthly Revenue")
plt.show()

print("="*70)
print("KPI ANALYSIS COMPLETED")
print("="*70)
```
