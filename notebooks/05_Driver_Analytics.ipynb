```
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option("display.max_columns",None)

# Load Dataset

df = pd.read_csv("../data/processed/final_delivery_dataset.csv")

print("="*70)
print("DRIVER PERFORMANCE ANALYSIS")
print("="*70)

# Driver Summary

driver = (

df.groupby("Delivery_person_ID")

.agg(

Orders=("ID","count"),

Revenue=("Revenue","sum"),

Average_Rating=("Delivery_person_Ratings","mean"),

Average_Time=("Time_taken(min)","mean"),

Efficiency=("Driver_Efficiency","mean"),

Driver_Score=("Driver_Score","mean")

)

.reset_index()

)

print(driver.head())

# Top Drivers by Revenue

top_revenue = (

driver

.sort_values(
"Revenue",
ascending=False
)

.head(10)

)

print(top_revenue)

plt.figure(figsize=(12,5))

plt.bar(
top_revenue["Delivery_person_ID"],
top_revenue["Revenue"]
)

plt.xticks(rotation=90)

plt.title("Top Drivers by Revenue")

plt.show()

# Best Rated Drivers

top_rating = (

driver

.sort_values(
"Average_Rating",
ascending=False
)

.head(10)

)

print(top_rating)

# Fastest Drivers

fastest = (

driver

.sort_values(
"Average_Time"
)

.head(10)

)

print(fastest)

# Most Efficient Drivers

efficient = (

driver

.sort_values(
"Efficiency",
ascending=False
)

.head(10)

)

print(efficient)


# Driver Score Distribution

plt.figure(figsize=(8,5))

plt.hist(
driver["Driver_Score"],
bins=25
)

plt.title("Driver Score Distribution")

plt.show()

# Driver Ratings Distribution

plt.figure(figsize=(8,5))

plt.hist(
driver["Average_Rating"],
bins=20
)

plt.title("Average Driver Ratings")

plt.show()

# Orders Completed

orders = (

driver

.sort_values(
"Orders",
ascending=False
)

.head(10)

)

plt.figure(figsize=(12,5))

plt.bar(
orders["Delivery_person_ID"],
orders["Orders"]
)

plt.xticks(rotation=90)

plt.title("Top Drivers by Orders")

plt.show()

# Driver Leaderboard

leaderboard = (

driver

.sort_values(
["Driver_Score","Revenue"],
ascending=False
)

)

leaderboard["Rank"] = range(
1,
len(leaderboard)+1
)

print(leaderboard.head(20))

leaderboard.to_csv(
"../outputs/driver_leaderboard.csv",
index=False
)

# Summary Statistics

print("\nSummary Statistics")

print(driver.describe())

print("="*70)
print("DRIVER ANALYTICS COMPLETED")
print("="*70)
```
